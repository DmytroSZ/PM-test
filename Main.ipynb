```python
query = """
        SELECT device.browser AS device_browser,
           sum(totals.visits) as total_visits,
           sum(totals.hits) as total_hits,
           sum(totals.pageviews) as total_pageviews
        FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
        GROUP BY device_browser
        ORDER BY total_visits  DESC,total_hits  DESC ,total_pageviews  DESC
        """

# Run the query, and return a pandas DataFrame
result = client.query(query).result().to_dataframe()
result
```

```python
query = """
       SELECT device.browser AS device_browser,
        geoNetwork.continent as continent ,
        geoNetwork.subContinent as subContinent,
           sum(totals.visits) as total_visits,
           sum(totals.hits) as total_hits,
           sum(totals.pageviews) as total_pageviews
        FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
        GROUP BY device_browser,continent, subContinent
        ORDER BY total_visits  DESC,total_hits  DESC ,total_pageviews  DESC
        """

# Run the query, and return a pandas DataFrame
result = client.query(query).result().to_dataframe()
result['device_browser']=np.where(np.isin(result['device_browser'] ,['Chrome','Safari','Firefox'])  , result['device_browser'], 'other')

aggr_browser_list = result.groupby('device_browser').agg({'total_visits':sum}).sort_values('total_visits',ascending=True).reset_index()                         
aggr_browser_list=aggr_browser_list['device_browser'].to_list()
browser_dict= dict(zip(aggr_browser_list, range(len(aggr_browser_list))))

result=result.groupby(['device_browser','continent']).sum().reset_index()
result['DB_Rank'] = result['device_browser'].map(browser_dict)
result=result.sort_values(by=['DB_Rank','total_visits'], ascending=False)
result = result.drop('DB_Rank', axis=1)
result
```

```python
query = """
       select 
        count( fullVisitorId ) as Unique_Visitors_with_second_hits_count,
        count(case when mn=1 then fullVisitorId end ) as Unique_New_Visitors_with_second_hits_count,
        count(case when mn<>1 then fullVisitorId end ) as Unique_Old_Visitors_with_second_hits_count
       from 
       (SELECT 
        fullVisitorId,
       min (visitNumber) as mn,
       max (visitNumber) as mx
          
        FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
        group by fullVisitorId
        ) a
        where mn<>mx 
        """

# Run the query, and return a pandas DataFrame
result = client.query(query).result().to_dataframe()
result
```

```python
query = """
        

select 
a.channelGrouping,
a.source,
a.Visits_count,
b.Visitor_with_revenue,
b.total_revenue,
b.total_revenue/b.Visitor_with_revenue as avg_revenue,
b.Visitor_with_revenue/a.Visits_count *100 as percent_Visitor_with_revenue,
b.total_revenue/a.Visits_count as Revenue_for_one_visitor

from 
(select 
channelGrouping,
trafficSource.source,
 sum(totals.visits) as Visits_count
 
 from 
 `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
 group by channelGrouping,source
 ) a left join
(SELECT 
    channelGrouping,
    source,
    COUNT(DISTINCT fullvisitorId) AS Visitor_with_revenue,
    SUM(total_revenue/1000000) AS total_revenue
  
FROM (
    SELECT 
        channelGrouping ,
        fullvisitorId,
        trafficSource.source as source,
        hits.transaction.transactionRevenue AS total_revenue 
FROM 
    `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`, 
    UNNEST (hits) AS hits
    where hits.transaction.transactionRevenue<>0
)
GROUP BY channelGrouping,source
)  b   on (a.channelGrouping=b.channelGrouping and a.source=b.source)   
order by 5 desc,3 desc
        """

# Run the query, and return a pandas DataFrame
result = client.query(query).result().to_dataframe()
result
```

```python
query = """

SELECT 
    product.v2ProductCategory AS product_category,
    sum(product.productRevenue/1000000) AS total_revenue,
    sum(product.productQuantity) as total_productQuantity,
    sum(product.productRevenue/1000000)/sum(product.productQuantity) as avg_revenue
FROM 
    `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`, 
    UNNEST (hits) AS hits,
    UNNEST(hits.product) AS product
    
WHERE 
   hits.transaction.transactionRevenue<>0
   and product.productQuantity > 0 
   
group by product_category
order by 2 desc
      

  """

# Run the query, and return a pandas DataFrame
result = client.query(query).result().to_dataframe()
result
```

```python
query = """

SELECT 
    product.v2ProductCategory as product_category,
    geoNetwork.region as region  ,
    sum(product.productRevenue/1000000) AS total_revenue
    
FROM 
    `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`, 
    UNNEST (hits) AS hits,
    UNNEST(hits.product) AS product
    
WHERE 
   hits.transaction.transactionRevenue<>0
   and product.productQuantity > 0 
   
group by product_category,region
order by 2 desc
      

  """

# Run the query, and return a pandas DataFrame
result = client.query(query).result().to_dataframe().replace({'region': {'not available in demo dataset': 'other'}})

result_pivot = pd.pivot_table(result, values='total_revenue', index=['region'],
                    columns=['product_category'], aggfunc=np.sum, fill_value=0, margins=True)
result_pivot=result_pivot.sort_values('All', ascending=False).drop('All', axis=1).sort_values('All', ascending=False, axis=1).sort_values('All', ascending=False, axis=1) .drop('All') 

ax = plt.pyplot.subplots(figsize=(16, 12))
ax=sns.heatmap(result_pivot, linewidths=.5, annot=True, fmt='.1f', cmap="YlGnBu", mask=(result_pivot==0))
```
