```python
query = """
        SELECT device.browser AS device_browser,
           sum(totals.visits) as total_visits,
           sum(totals.hits) as total_hits,
           sum(totals.pageviews) as total_pageviews
        FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
        GROUP BY device_browser
        ORDER BY total_visits  DESC,total_hits  DESC ,total_pageviews  DESC
        """

# Run the query, and return a pandas DataFrame
result = client.query(query).result().to_dataframe()
result
```

```python
query = """
       SELECT device.browser AS device_browser,
        geoNetwork.continent as continent ,
        geoNetwork.subContinent as subContinent,
           sum(totals.visits) as total_visits,
           sum(totals.hits) as total_hits,
           sum(totals.pageviews) as total_pageviews
        FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
        GROUP BY device_browser,continent, subContinent
        ORDER BY total_visits  DESC,total_hits  DESC ,total_pageviews  DESC
        """

# Run the query, and return a pandas DataFrame
result = client.query(query).result().to_dataframe()
result['device_browser']=np.where(np.isin(result['device_browser'] ,['Chrome','Safari','Firefox'])  , result['device_browser'], 'other')

aggr_browser_list = result.groupby('device_browser').agg({'total_visits':sum}).sort_values('total_visits',ascending=True).reset_index()                         
aggr_browser_list=aggr_browser_list['device_browser'].to_list()
browser_dict= dict(zip(aggr_browser_list, range(len(aggr_browser_list))))

result=result.groupby(['device_browser','continent']).sum().reset_index()
result['DB_Rank'] = result['device_browser'].map(browser_dict)
result=result.sort_values(by=['DB_Rank','total_visits'], ascending=False)
result = result.drop('DB_Rank', axis=1)
result
```

```python
query = """
        SELECT 
        sum(totals.visits) as Visits_count,
        sum(case when visitNumber=1 then totals.visits else 0 end) as Visits_by_New_Visitors_count,
        sum(case when visitNumber=1 then 0 else totals.visits end) as Visits_by_Old_Visitors_count,
        count(distinct fullVisitorId) as Unique_Visitor_count,
        count(distinct case when visitNumber=1 then fullVisitorId  end) as Unique_New_Visitor_count,
        count(distinct fullVisitorId)-count(distinct case when visitNumber=1 then fullVisitorId  end) as Unique_Old_Visitor_count
          
        FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
        
        """
'''Other way
sum(totals.newvisits) as Visits_by_New_Visitors_count '''
# Run the query, and return a pandas DataFrame
result = client.query(query).result().to_dataframe()
result
```

```python
plt.pyplot.figure (figsize=(5,5))
labels = 'Visits_by_New_Visitors_count', 'Visits_by_Old_Visitors_count'
sizes = [result['Visits_by_New_Visitors_count'].values[0],result['Visits_by_Old_Visitors_count'].values[0]]
plt.pyplot.pie(sizes, labels=labels, autopct='%1.1f%%',startangle=90)
plt.pyplot.show()
```

```python
query = """
       select 
        count( fullVisitorId ) as Unique_Visitors_with_second_hits_count,
        count(case when mn=1 then fullVisitorId end ) as Unique_New_Visitors_with_second_hits_count,
        count(case when mn<>1 then fullVisitorId end ) as Unique_Old_Visitors_with_second_hits_count
       from 
       (SELECT 
        fullVisitorId,
       min (visitNumber) as mn,
       max (visitNumber) as mx
          
        FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
        group by fullVisitorId
        ) a
        where mn<>mx 
        """

# Run the query, and return a pandas DataFrame
result = client.query(query).result().to_dataframe()
result
```

```python
plt.pyplot.figure (figsize=(5,5))
labels = 'Unique_New_Visitors_with_second_hits_count', 'Unique_Old_Visitors_with_second_hits_count'
sizes = [result['Unique_New_Visitors_with_second_hits_count'].values[0],result['Unique_Old_Visitors_with_second_hits_count'].values[0]]
plt.pyplot.pie(sizes, labels=labels, autopct='%1.1f%%',startangle=90)
plt.pyplot.show()
```

```python
query = """
        select totals.newvisits,
        sum(totals.pageviews)/sum(totals.visits) as Avg_pageviews,
        sum(totals.hits)/sum(totals.visits) as Avg_hits
        FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`
        group by totals.newvisits
        """

# Run the query, and return a pandas DataFrame
result = client.query(query).result().to_dataframe().fillna('Old_visits').replace({'newvisits': {1: 'New_visits'}})
result
```
